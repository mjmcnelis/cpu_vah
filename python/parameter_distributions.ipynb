```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#[[b, N, p, w, dmin, sigmak, Tsw, Tk, etask, aL, aH, zetasN, Tp, wz, lambdaz]]

data = np.loadtxt('random_model_parameters/model_parameters_1.dat')
data = data.reshape(1, len(data))

for n in range(1, 1000):
    sample = np.loadtxt('random_model_parameters/model_parameters_' + str(n + 1) + '.dat')
    sample = sample.reshape(1, len(sample))
    data = np.concatenate((data, sample))
       
print(data.shape)

b = data[:,0]
N = data[:,1]

plt.figure(figsize = (6,4))
plt.hist(b, bins = 10)
plt.title("Impact parameter",fontsize = 15)
plt.xlabel("b [fm]",fontsize = 15)
plt.ylabel("counts",fontsize = 15)
plt.show()

plt.hist(N, bins = 10)
plt.title("Normalization",fontsize = 15)
plt.xlabel("N [GeV]",fontsize = 15)
plt.ylabel("counts",fontsize = 15)
plt.show()
```
